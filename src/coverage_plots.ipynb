```python
import glob
import os
import pandas as pd
import matplotlib.pyplot as plt

def read_coverage(filename):
    df = pd.read_csv(filename, sep="\t", header=None, names=["chrom", "pos", "coverage"])
    return df["pos"], df["coverage"]

# Get coverage files for each segment.
files_S = glob.glob("/Users/ricardorivero/Documents/GitHub/SNVler/Results/Reports/*_S.coverage")
files_M = glob.glob("/Users/ricardorivero/Documents/GitHub/SNVler/Results/Reports/*_M.coverage")
files_L = glob.glob("/Users/ricardorivero/Documents/GitHub/SNVler/Results/Reports/*_L.coverage")

# Plot for Segment S
plt.figure(figsize=(10, 6))
for filename in files_S:
    pos, coverage = read_coverage(filename)
    sample_name = os.path.basename(filename).split(".")[0]
    plt.plot(pos, coverage, label=sample_name)
plt.title("Coverage Plot for Segment S")
plt.xlabel("Genomic Position")
plt.ylabel("Coverage")
plt.legend()
plt.show()

# Plot for Segment M
plt.figure(figsize=(10, 6))
for filename in files_M:
    pos, coverage = read_coverage(filename)
    sample_name = os.path.basename(filename).split(".")[0]
    plt.plot(pos, coverage, label=sample_name)
plt.title("Coverage Plot for Segment M")
plt.xlabel("Genomic Position")
plt.ylabel("Coverage")
plt.legend()
plt.show()

# Plot for Segment L
plt.figure(figsize=(10, 6))
for filename in files_L:
    pos, coverage = read_coverage(filename)
    sample_name = os.path.basename(filename).split(".")[0]
    plt.plot(pos, coverage, label=sample_name)
plt.title("Coverage Plot for Segment L")
plt.xlabel("Genomic Position")
plt.ylabel("Coverage")
plt.legend()
plt.show()
```
